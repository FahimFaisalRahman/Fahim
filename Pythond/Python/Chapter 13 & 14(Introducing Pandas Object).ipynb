**Series as Specialized Dictionary**

```python
population_dict = {
    'California': 3958223,
    'Texas': 29145505,
    'Florida': 2158187,
    'New York': 20201249,
    'Pennsylvania': 13002700}

population= pd.Series(population_dict)
```

```python
population
```

```python
population['California']
```

```python
area_dict = {
    'California': 423967,
    'Texas': 695662,
    'Florida': 170312, 
    'New York': 141297,
    'Pennsylvania': 119280}

area = area_dict
```

```python
states = pd.DataFrame({'population': population, 
                       'area': area})
states
```

```python
states = pd.DataFrame({
    'Population': [3958223, 29145505, 2158187, 20201249, 13002700],
    'Area': [423967, 695662, 170312, 141297, 119280],}, 
    index = ['California', 'Texas', 'Florida', 'New York', 'Penssylvania'])
```
