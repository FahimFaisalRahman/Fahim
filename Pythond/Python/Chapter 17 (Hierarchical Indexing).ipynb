```python
import pandas as pd

# Define the index and population data
index = [('California', 2010), ('California', 2020), 
         ('New York', 2010), ('New York', 2020), 
         ('Texas', 2010), ('Texas', 2020)]

population = [37253956, 39538223,
              9378102, 20201249,
              25145561, 29145505]

# Create the MultiIndex
index = pd.MultiIndex.from_tuples(index, names=['State', 'Year'])

# Create the pop DataFrame
pop = pd.DataFrame(population, index=index, columns=['total'])

# Create pop_df with the total column as a Series
pop_df = pd.DataFrame({
    'total': pop['total'],  # Extract the 'total' column as a Series
    'under18': [9284094, 8898092, 4318033, 4181528, 6879014, 7432474]
}, index=index)

# Display the result
print(pop_df)
```

```python
f_u18 = pop_df['under18'] / pop_df['total']
```

```python
print(f_u18.unstack())
```
