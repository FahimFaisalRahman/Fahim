### Modifying, Adding, and Removing Elements 

**Modifying Elements in a List**

```python
motorcycles = ['honda', 'yamaha', 'suzuki']
print(motorcycles)
```

```python
motorcycles[0] = 'ducati'
print(motorcycles)
```

**Adding Elemnts to a list**

*Appending Elements to the End of a List*

```python
motorcycles.append('ducati')
```

```python
print(motorcycles)
```

```python
motorcycles = []
motorcycles.append('honda')
motorcycles.append('yamaha')
motorcycles.append('suzuki')
```

```python
print(motorcycles)
```

*Inserting Elements into a List*

```python
motorcycles = ['honda', 'yamaha', 'suzuki']
motorcycles.insert(0, 'ducati')
print(motorcycles)
```

**Removing Elements from a list**

*Removing an Item Using the del Statement* 

```python
motorcycles = ['honda', 'yamaha', 'suzuki']
print(motorcycles)
```

```python
del motorcycles [0]
print(motorcycles)
```

*Removing Items usings pop() method*

```python
popped_motorcycles = motorcycles.pop()
```

```python
print(popped_motorcycles)
```

```python
last_owned_motorcycles = popped_motorcycles
```

```python
print(f"The last motorcycles I owned was a {last_owned_motorcycles.title()}.")
```

```python
first_owned_motorcycles = motorcycles.pop(0)
```

```python
print(f"The first motorcycle I owned was a {first_owned_motorcycles.title()}.")
```

```python
print(motorcycles)
```

**Sorting a List Permanently with the sort() Method**

```python
cars = ['bmw', 'audi', 'toyota', 'subaru']
cars.sort()
print(cars)
```

```python
cars.sort(reverse=True)
```

```python
print(cars)
```

```python
cars.reverse()
print(cars)
```
