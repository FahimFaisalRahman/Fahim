```python
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    'numbers': [1, 2, 3],
    'floats': [1.5, 2.5, 3.5],
    'names': ['Alice', 'Bob', 'Charlie']
})

# New row to append
new_row = pd.DataFrame([{'numbers': 100, 'floats': 5.75, 'names': 'Jill'}])

# Append the new row using concat
df = pd.concat([df, new_row], ignore_index=True)

print(df)
```

```python
order = pd.DataFrame({
    'Order ID' : [1,2,3,4],
    'Customer Name': ['Jane', 'June','May', 'July']
}, index = ['a','b','c','d'])
```

```python
Customer = pd.DataFrame({
    'Customer Name': ['Jane', 'June', 'May', 'July'],
    'Country': ['USA', 'Bangladesh', 'Ethiopia','Nigeria']
}, index = ['a', 'b', 'c', 'd'])
```

```python
order = order.merge(Customer, how='left', left_on = 'Customer Name', right_on='Customer Name')
```
