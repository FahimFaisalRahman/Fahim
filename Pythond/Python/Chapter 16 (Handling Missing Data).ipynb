**Detecting Null Values**

```python
data = pd.Series([1, np.nan, 'hello', None])
```

```python
data.isnull()
```

**Dropping Null Values**

```python
data.dropna()
```

```python
df = pd.DataFrame([[1, np.nan, 2],
                   [2, 3, 5],
                   [np.nan, 4, 6]])
df
```

```python
df.dropna()
```

```python
df.dropna(axis="columns")
```

```python
df[3] = np.nan
df
```

```python
df.dropna(axis="columns", how="all")
```

**Filling Null Values**

```python
data = pd.DataFrame([[1, np.nan, 2, 3],
                     [2,3,np.nan, 3],
                      [1,2,3,np.nan]])
```

```python
data
```

```python
data.fillna(0)
```
