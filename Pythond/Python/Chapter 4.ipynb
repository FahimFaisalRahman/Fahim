# Working with Lists

**Looping Through on Entire List**

```python
magicians = ['alice', 'david', 'carolina']
for magician in magicians:
    print(magician)
```

```python
magicians = ['alice', 'david', 'carolina']
for magician in magicians:
    print(f"{magician.title()}, that's a great trick!")
```

```python
magicians = ['alice', 'david', 'carolina']
for magician in magicians:
    print(f"{magician.title()}, that's a great trick!")
    print(f"I can't wait to see your next magic trick, {magician.title()}.\n")
```

```python
animals = ['dolphins', 'sharks', 'ichthyosaurs']

for animal in animals:
    print(f"A {animal.title()}, would make a great pet.")

print('\nThese animals evolved a similar swimming due to convergent evolutions.')
```

**Looping Through a Slice**

```python
players = ['charles', 'martina', 'michael', 'florence', 'eli']

print("Here are the first three players on my team:")
for player in players[:3]:
    print(player.title())
```

```python
teams = ['Manchester United', 'Manchester City', 'Chelsea', 'Liverpool', 'Tottemham Hotspur', 'Leicester City']

print("Here are the worst three teams in EPL")
for team in teams[3:]:
    print(team.title())
```

**Copying in a List**

```python
my_foods = ['pizza', 'falafel', 'carrot cake']
friend_foods = my_foods[:]

print("My favorite foods are:")
print(my_foods)

print("\nMy Friend's favorite foods are:")
print(friend_foods)
```

```python
my_foods.append('cannoli')
friend_foods.append('ice cream')

print("My favorite foods are:")
print(my_foods)

print("\nMy Friend's favorite foods are:")
print(friend_foods)
```

```python
#4-10 Slices 
teams = ['Manchester City', 'Mancher United', 'Chelsea', 'Liverpool','Tottemham Hotspur','Leicester City', 'Arsenal', 'West Bromwhich', 'Crystal City']

print("The best three teams in EPL")
for team in teams[:3]:
    print(team.title())

print("\nThe middle three teams in EPL")
for team in teams[3:6]:
    print(team.title())

print("\nThe worst three teams in EPL")
for team in teams[6:9]:
    print(team.title())
```
